```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
train=pd.read_csv('fraudTrain.csv')
test=pd.read_csv('fraudTest.csv')
```

```python
train.head()
```

```python
test.head()
```

```python
# calculating latitudinal and longitudinal distance for training data
train['lat_dist']=abs(round(train['merch_lat']-train['lat'],2))
train['long_dist']=abs(round(train['merch_long']-train['long'],2))
train.head()

# calculating latitudinal and longitudinal distance for test data
test['lat_dist']=abs(round(test['merch_lat']-test['lat'],2))
test['long_dist']=abs(round(test['merch_long']-test['long'],2))
test.head()
```

```python
# converting in same unit the date of birth and transaction date for training data
train['DOB'] = pd.to_datetime(train['dob'])
train['Transaction_Date'] = pd.to_datetime(train['trans_date_trans_time'])


# calculating age in years for training data
train['age']=(train['Transaction_Date'] - train['DOB']).astype('<m8[ns]')

nanoseconds_per_year = 31557600000000000
train['years']= train['age']/np.timedelta64(nanoseconds_per_year,'ns')

train['years']=round(train['years'],0 )

```

```python
# converting in same unit the date of birth and transaction date for test data

test['DOB'] = pd.to_datetime(test['dob'])
test['Transaction_Date'] = pd.to_datetime(test['trans_date_trans_time'])


# calculating age in years for training data
test['age']=(test['Transaction_Date'] - test['DOB']).astype('<m8[ns]')

nanoseconds_per_year = 31557600000000000
test['years']= test['age']/np.timedelta64(nanoseconds_per_year,'ns')

test['years']=round(test['years'],0)
```

```python
# Replacing M and F of gender column with 0 and 1 respectively

train['gender']=train['gender'].replace({'M':0,'F':1})

test['gender']=test['gender'].replace({'M':0,'F':1})
```

```python
train=train[["amt",'gender','lat_dist','long_dist','years','is_fraud','city_pop']]

train.head()
```

```python
test=test[["amt",'gender','lat_dist','long_dist','years','is_fraud','city_pop']]
test.head()
```

```python
# splitting the data 

y_train=train['is_fraud']
x_train=train.drop('is_fraud',axis=1)

y_test=test['is_fraud']
x_test=test.drop('is_fraud',axis=1)
```

```python
from sklearn.linear_model import LogisticRegression

model=LogisticRegression().fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,y_pred))
```
